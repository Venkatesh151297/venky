```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',
'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
df
```

```python
df['FlightNumber']=[10045,10055,10065,10075,10085]
df.FlightNumber.astype(int)
df
```

```python
df[['From','To']]=df.From_To.str.split('_',expand=True)
df
```

```python
df.drop(labels='From_To',axis=1,inplace=True)
df
```

```python
df['From']=df.From.str.title()
df['To']=df.To.str.title()
df
```

```python
b=[]
for i in range(0,len(df)):
    a=len(df['RecentDelays'][i])
    b.append(a)
max(b)

```

```python
for i in range(0,len(df)):    
    if len(df['RecentDelays'][i]) < max(b):
        for j in range(max(b)-len(df['RecentDelays'][i])):
            df['RecentDelays'][i].append(np.nan)
df
```

```python
delays_1=[]
delays_2=[]
delays_3=[]
for i in range(0,5):
    a=df['RecentDelays'][i][0]
    delays_1.append(a)
for i in range(0,5):
    a=df['RecentDelays'][i][1]
    delays_2.append(a)
for i in range(0,5):
    a=df['RecentDelays'][i][2]
    delays_3.append(a)
    
```

```python
delays=pd.DataFrame([delays_1,delays_2,delays_3])
delays1=delays.T
delays1.col
```

```python
delays1.set_axis(['delay1','delay2','delay3'], axis=1)
```

```python
delays1
```

```python
frames = [df, delays1]
  
result = pd.concat(frames,axis=1)
result
```

```python
result.drop('RecentDelays',axis=1,inplace=True)
```
